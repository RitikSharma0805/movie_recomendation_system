```python
rating=pd.DataFrame(df.groupby('title').mean()['rating'])
```

```python
rating['num of rating']=pd.DataFrame(df.groupby('title').count()['rating'])
```

```python
rating.sort_values(by='rating',ascending=False)
```

```python
movie_mat=df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
movie_mat
```

```python
rating.sort_values('num of rating',ascending=False).head()
```

```python
movie_user_rating=movie_mat['3 Idiots (2009)']
```

```python
similar=movie_mat.corrwith(movie_user_rating)
```

```python
corr_movie=pd.DataFrame(similar,columns=['correlation'])
```

```python
corr_movie.dropna(inplace=True)
```

```python
corr_movie
```

```python
corr_movie.sort_values('correlation',ascending=False).head()
```

```python
rating.head()
print(type(corr_movie))
```

```python
corr_movie=corr_movie.join(rating['num of rating'])
```

```python
corr_movie[corr_movie['num of rating']>100].sort_values('correlation',ascending=False)
```

```python
def predict_movie(movie_name):
    movie_user_rating=movie_mat[movie_name]
    similar_to_movie=movie_mat.corrwith(movie_user_rating)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(rating['num of rating'])
    prediction=corr_movie[corr_movie['num of rating']>100].sort_values('correlation',ascending=False)
   
    return prediction
```
